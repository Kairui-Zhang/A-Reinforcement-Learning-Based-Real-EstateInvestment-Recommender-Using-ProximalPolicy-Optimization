```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Load Data ---
# Load the two local Excel files
all_props_df = pd.read_excel('all_properties.xlsx')
price_rent_df = pd.read_excel('price_rent_index.xlsx')

# --- 2. Set Plotting Style ---
# Use seaborn's aesthetic style
sns.set_theme(style="whitegrid")

# --- 3. Start Plotting ---

# Figure 1: Distribution of Core Numerical Features (Price and Rent)
print("Generating Figure 1: Distribution of Core Numerical Features...")
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Price Distribution Histogram
sns.histplot(all_props_df['price'], bins=50, kde=True, ax=axes[0])
axes[0].set_title('Distribution of Price', fontsize=16)
axes[0].set_xlabel('Price (EUR)', fontsize=12)
axes[0].set_ylabel('Frequency', fontsize=12)

# Base Rent Distribution Histogram
sns.histplot(all_props_df['base_rent'], bins=50, kde=True, ax=axes[1], color='orange')
axes[1].set_title('Distribution of Base Rent', fontsize=16)
axes[1].set_xlabel('Rent (EUR)', fontsize=12)
axes[1].set_ylabel('Frequency', fontsize=12)

plt.suptitle('Distribution of Core Numerical Features', fontsize=20, y=1.02)
plt.tight_layout()
plt.show()

# Figure 2: House Price Distribution by Federal State
print("Generating Figure 2: House Price Distribution by Federal State...")
plt.figure(figsize=(18, 10))
# Focus on the top 10 states with the most listings
top_states = all_props_df['obj_regio1'].value_counts().nlargest(10).index

sns.boxplot(data=all_props_df[all_props_df['obj_regio1'].isin(top_states)], x='price', y='obj_regio1', orient='h')
plt.title('Comparison of House Price Distribution in Major Federal States', fontsize=20)
plt.xlabel('Price (EUR)', fontsize=14)
plt.ylabel('Federal State (obj_regio1)', fontsize=14)
# Use a log scale for the x-axis for better clarity due to the wide price range
plt.xscale('log')
plt.grid(True, which="both", ls="--")
plt.show()

# Figure 3: Correlation Heatmap of Key Numerical Variables
print("Generating Figure 3: Correlation Heatmap of Key Numerical Variables...")
corr_cols = ['price', 'base_rent', 'service_charge', 'obj_lotArea', 'obj_yearConstructed', 'GDP_Per_Capita', 'Population_State']
corr_matrix = all_props_df[corr_cols].corr()

plt.figure(figsize=(12, 9))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Heatmap of Key Numerical Variables', fontsize=20)
plt.show()

# Figure 4: Price Analysis by House Type
print("Generating Figure 4: Price Analysis by House Type...")
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Count of House Types
sns.countplot(data=all_props_df, x='house_type', ax=axes[0], palette='viridis')
axes[0].set_title('Count of House Types', fontsize=16)
axes[0].set_xlabel('House Type', fontsize=12)
axes[0].set_ylabel('Count', fontsize=12)

# Boxplot of Prices by House Type
sns.boxplot(data=all_props_df, x='house_type', y='price', ax=axes[1], palette='viridis')
axes[1].set_title('Price Distribution by House Type', fontsize=16)
axes[1].set_xlabel('House Type', fontsize=12)
axes[1].set_ylabel('Price (EUR)', fontsize=12)
# Also use a log scale for the y-axis
axes[1].set_yscale('log')

plt.suptitle('Price Analysis by House Type', fontsize=20, y=1.02)
plt.tight_layout()
plt.show()

# Figure 5: Annual Growth Trend of House Price and Rent Indices
print("Generating Figure 5: Annual Growth Trend of House Price and Rent Indices...")
plt.figure(figsize=(18, 9))

# Melt the dataframe for easier plotting with seaborn
price_rent_melted = price_rent_df.melt(id_vars=['Year'], var_name='Index_Type', value_name='Growth_Rate')

sns.lineplot(data=price_rent_melted, x='Year', y='Growth_Rate', hue='Index_Type', marker='o')
plt.title('Annual Growth Trend of House Price and Rent Indices', fontsize=20)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Annual Growth Rate', fontsize=14)
# Baseline (No Growth at 1.0)
plt.axhline(1.0, color='red', linestyle='--', label='Baseline (No Growth at 1.0)')
plt.legend(title='Index Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, which="both", ls="--")
plt.show()

print(" All EDA charts have been generated successfully!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# --- 1. Load Data ---
# Load your local Excel file
try:
    all_props_df = pd.read_excel('all_properties.xlsx')
except FileNotFoundError:
    print("Error: 'all_properties.xlsx' not found. Please make sure the file is in the correct directory.")
    exit()

# --- 2. Set Plotting Style ---
sns.set_theme(style="whitegrid")

# --- 3. Bin the data and create plots ---
print("Generating binned distribution plots for Price and Total Rent...")

# Create a figure with two subplots, side by side
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

# --- Plot for 'price' with custom bins ---
# Define the bins for price as per your request
price_bins = [0, 30000, 100000, 200000, 500000, 1000000, np.inf]
price_labels = ['0-30k', '30k-100k', '100k-200k', '200k-500k', '500k-1M', '>1M']

# Create a new column with the binned data
all_props_df['price_bin'] = pd.cut(all_props_df['price'], bins=price_bins, labels=price_labels, right=False)

# Create the bar plot
price_counts = all_props_df['price_bin'].value_counts().sort_index()
sns.barplot(x=price_counts.index, y=price_counts.values, ax=axes[0], palette="viridis")
axes[0].set_title('Distribution of Price in Custom Bins', fontsize=16)
axes[0].set_xlabel('Price Range (EUR)', fontsize=12)
axes[0].set_ylabel('Number of Properties (Count)', fontsize=12)
# Add count labels on top of bars
for index, value in enumerate(price_counts):
    axes[0].text(index, value + (price_counts.max() * 0.01), str(value), ha='center', va='bottom')


# --- Plot for 'obj_totalRent' with custom bins ---
# Check if 'obj_totalRent' column exists
if 'obj_totalRent' in all_props_df.columns:
    # Define the bins for total rent as per your request
    rent_bins = [0, 200, 800, 1500, 3000, np.inf]
    rent_labels = ['0-200', '201-800', '801-1500', '1501-3k', '>3k']

    # Create a new column with the binned data
    all_props_df['rent_bin'] = pd.cut(all_props_df['obj_totalRent'], bins=rent_bins, labels=rent_labels, right=False)

    # Create the bar plot
    rent_counts = all_props_df['rent_bin'].value_counts().sort_index()
    sns.barplot(x=rent_counts.index, y=rent_counts.values, ax=axes[1], palette="plasma")
    axes[1].set_title('Distribution of Total Rent in Custom Bins', fontsize=16)
    axes[1].set_xlabel('Total Rent Range (EUR)', fontsize=12)
    axes[1].set_ylabel('Number of Properties (Count)', fontsize=12)
    # Add count labels on top of bars
    for index, value in enumerate(rent_counts):
        axes[1].text(index, value + (rent_counts.max() * 0.01), str(value), ha='center', va='bottom')
else:
    axes[1].text(0.5, 0.5, "'obj_totalRent' column not found", ha='center', va='center', fontsize=12)
    axes[1].set_title('Distribution of Total Rent', fontsize=16)


# Add a main title and adjust layout
plt.suptitle('Binned Data Distribution for Price and Total Rent', fontsize=20, y=1.02)
plt.tight_layout()
plt.show()

print("\n Binned plot generation complete!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# --- 1. Load Data ---
# Load your local Excel file
try:
    all_props_df = pd.read_excel('all_properties.xlsx')
except FileNotFoundError:
    print("Error: 'all_properties.xlsx' not found. Please make sure the file is in the correct directory.")
    exit()

# --- 2. Set Plotting Style ---
sns.set_theme(style="whitegrid")

# --- 3. Bin the data ---
# Define the bins for price
price_bins = [0, 30000, 100000, 200000, 500000, 1000000, np.inf]
price_labels = ['0-30k', '30k-100k', '100k-200k', '200k-500k', '500k-1M', '>1M']
all_props_df['price_bin'] = pd.cut(all_props_df['price'], bins=price_bins, labels=price_labels, right=False)

# Define the bins for total rent (if the column exists)
if 'obj_totalRent' in all_props_df.columns:
    rent_bins = [0, 200, 800, 1500, 3000, np.inf]
    rent_labels = ['0-200', '201-800', '801-1500', '1501-3k', '>3k']
    all_props_df['rent_bin'] = pd.cut(all_props_df['obj_totalRent'], bins=rent_bins, labels=rent_labels, right=False)

# --- 4. Generate Plot 1: Price Distribution ---
print("Generating Plot 1: Binned Distribution of Price...")
plt.figure(figsize=(12, 7))
price_counts = all_props_df['price_bin'].value_counts().sort_index()
sns.barplot(x=price_counts.index, y=price_counts.values, palette="viridis")
plt.title('Distribution of Price', fontsize=18)
plt.xlabel('Price Range (EUR)', fontsize=14)
plt.ylabel('Number of Properties (Count)', fontsize=14)
# Add count labels on top of bars
for index, value in enumerate(price_counts):
    plt.text(index, value + (price_counts.max() * 0.01), str(value), ha='center', va='bottom', fontsize=11)
plt.tight_layout()
plt.show()

# --- 5. Generate Plot 2: Total Rent Distribution ---
if 'obj_totalRent' in all_props_df.columns:
    print("\nGenerating Plot 2: Binned Distribution of Total Rent...")
    plt.figure(figsize=(12, 7))
    rent_counts = all_props_df['rent_bin'].value_counts().sort_index()
    sns.barplot(x=rent_counts.index, y=rent_counts.values, palette="plasma")
    plt.title('Distribution of Total Rent', fontsize=18)
    plt.xlabel('Total Rent Range (EUR)', fontsize=14)
    plt.ylabel('Number of Properties (Count)', fontsize=14)
    # Add count labels on top of bars
    for index, value in enumerate(rent_counts):
        plt.text(index, value + (rent_counts.max() * 0.01), str(value), ha='center', va='bottom', fontsize=11)
    plt.tight_layout()
    plt.show()
else:
    print("\nSkipping Total Rent plot: 'obj_totalRent' column not found.")

print("\n All plots generated successfully!")
```
